```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import os

# -----------------------------
# Load Data
# -----------------------------
df = pd.read_csv("/content/sample_data/sales_data (4).csv")

# Convert Date column if exists
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])

# Create visualization folder
if not os.path.exists("visualizations"):
    os.makedirs("visualizations")

# -----------------------------
# Styling
# -----------------------------
sns.set_theme(style="whitegrid")
sns.set_palette("viridis")

# -----------------------------
# 1️⃣ BOX PLOT
# -----------------------------
plt.figure(figsize=(8,6))
sns.boxplot(x='Product', y='Total_Sales', data=df) # Changed 'Sales' to 'Total_Sales'
plt.title('Sales Distribution by Product') # Changed 'Category' to 'Product'
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("visualizations/boxplot.png")
plt.close()

# -----------------------------
# 2️⃣ VIOLIN PLOT
# -----------------------------
plt.figure(figsize=(8,6))
sns.violinplot(x='Product', y='Total_Sales', data=df) # Changed 'Sales' to 'Total_Sales'
plt.title('Sales Density by Product') # Changed 'Category' to 'Product'
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("visualizations/violinplot.png")
plt.close()

# -----------------------------
# 3️⃣ CORRELATION HEATMAP
# -----------------------------
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("visualizations/heatmap.png")
plt.close()

# -----------------------------
# 4️⃣ STATIC 2x2 DASHBOARD
# -----------------------------
fig, axes = plt.subplots(2,2, figsize=(14,10))

sns.boxplot(x='Product', y='Total_Sales', data=df, ax=axes[0,0]) # Changed 'Sales' to 'Total_Sales'
axes[0,0].set_title("Sales Distribution by Product") # Changed 'Category' to 'Product'

sns.countplot(x='Region', data=df, ax=axes[0,1]) # Changed 'Customer_Segment' to 'Region'
axes[0,1].set_title("Customer Segments by Region") # Updated title

sns.barplot(x='Product', y='Total_Sales', data=df, ax=axes[1,0]) # Changed 'Sales' to 'Total_Sales'
axes[1,0].set_title("Average Sales by Product") # Changed 'Category' to 'Product'

sns.histplot(df['Total_Sales'], kde=True, ax=axes[1,1]) # Changed 'Sales' to 'Total_Sales'
axes[1,1].set_title("Sales Histogram")

plt.tight_layout()
plt.savefig("visualizations/dashboard_static.png")
plt.close()

# -----------------------------
# 5️⃣ INTERACTIVE PLOTLY BAR CHART
# -----------------------------
interactive_bar = px.bar(
    df,
    x="Product", # Changed 'Category' to 'Product'
    y="Total_Sales", # Changed 'Sales' to 'Total_Sales'
    color="Region", # Changed 'Customer_Segment' to 'Region'
    hover_data=df.columns,
    title="Interactive Sales by Product and Region" # Updated title
)

interactive_bar.write_html("visualizations/interactive_bar.html")

# -----------------------------
# 6️⃣ INTERACTIVE LINE CHART (Sales Trend)
# -----------------------------
if "Date" in df.columns:
    trend = px.line(
        df,
        x="Date",
        y="Total_Sales", # Changed 'Sales' to 'Total_Sales'
        color="Product", # Changed 'Category' to 'Product'
        title="Sales Trend Over Time"
    )
    trend.write_html("visualizations/sales_trend.html")

# -----------------------------
# 7️⃣ ADVANCED INTERACTIVE DASHBOARD
# -----------------------------
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Sales by Product", # Changed 'Category' to 'Product'
                                    "Region Distribution", # Changed 'Segment Distribution' to 'Region Distribution'
                                    "Sales Histogram",
                                    "Sales Trend"),
                    specs=[[{}, {'type': 'domain'}], [{}, {}]]) # Added specs for pie chart

# Bar
fig.add_trace(
    go.Bar(x=df["Product"], y=df["Total_Sales"]),
    row=1, col=1
)

# Segment
segment_counts = df["Region"].value_counts()
fig.add_trace(
    go.Pie(labels=segment_counts.index,
           values=segment_counts.values),
    row=1, col=2
)

# Histogram
fig.add_trace(
    go.Histogram(x=df["Total_Sales"]),
    row=2, col=1
)

# Trend if available
if "Date" in df.columns:
    fig.add_trace(
        go.Scatter(x=df["Date"], y=df["Total_Sales"], mode='lines'),
        row=2, col=2
    )

fig.update_layout(height=800, width=1000,
                  title_text="Interactive Sales Dashboard")

fig.write_html("visualizations/full_dashboard.html")

print("✅ Dashboard created successfully!")
```
